### Cognition-relevant brain spatiotemporal dynamic pattern

* The top-weighted time-series feature in the cognition mode is **VarRiotest_stats** (computed by function *[SY_VarRatioTest](https://github.com/benfulcher/hctsa/blob/main/Operations/SY_VarRatioTest.m)* from hcsta toolbox.

* Utilizing data from HCP-YA subjects incorporated into the CCA analysis, we identified significant brain map based on the correlation between individual total cognition composition scores and whole-brain VarRiotest_stats features.


```python
feature_mat = np.load('../data/time_series_matrix/datawb_vartiotest.npy')
behavior_df = pd.read_csv('../data/HCPYA_behavior/CCA_behavior_y.csv')
beav_x = behavior_df['CogTotalComp_Unadj']
```

```python
_corr = CCAHCP.get_feature_behavior_corr(feature_mat.mean(axis = 2) ,beav_x)
cog_spatial_pattern = CCAHCP.viz_xdata(_corr['r_sig'],colormap ='bwr_r',title = 'CogMode')  
```
